```python
# import the dataset
path=r'C:\Users\admin\Desktop\Draft\Final_dataset.xlsx'
data=pd.read_excel(path)

static_columns=['Company Name','Industry','comser_dummy','condis_dummy','consta_dummy',
                'energy_dummy','financials_dummy','health_dummy','industrials_dummy',
                'inftec_dummy','materials_dummy','realestate_dummy']

time_varying_columns=[col for col in data.columns if col not in static_columns]
```

```python
# convert to long format
data_long=pd.melt(data,
                  id_vars=static_columns,
                  value_vars=time_varying_columns,
                  var_name='Variable',
                  value_name='Value')

data_long['Year']=data_long['Variable'].str.extract(r'(\d{4})')
data_long['Variable']=data_long['Variable'].str.replace(r'_\d{4}','',regex=True)
```

```python
panel_data['Year']=panel_data['Year'].astype(int)

panel_data=panel_data.set_index(['Company Name', 'Year'])
```

```python
# add interaction term between ESG and Year 
panel_data['Year']=panel_data.index.get_level_values('Year')

year_dummies=pd.get_dummies(panel_data['Year'],prefix='Year',drop_first=True)
panel_data=panel_data.join(year_dummies)

for year_dummy in year_dummies.columns:
    panel_data[f'esg_{year_dummy}_interaction']=panel_data['esg']*panel_data[year_dummy]

x_int=panel_data[[f'esg_{col}_interaction' for col in year_dummies.columns]+['age','lev','beta','comser_dummy','condis_dummy','consta_dummy',
                                                                             'financials_dummy','health_dummy','industrials_dummy',
                                                                             'inftec_dummy','materials_dummy','realestate_dummy']]
y_int=panel_data['TQ']
```

```python
# Impact of ESG on TQ with interaction terms
model=RandomEffects(y_int,x_int).fit(cov_type='robust')
model.summary
```

```python
# extract model coefficients
coefficients=model.params
coefficients
```

```python
# Plot interaction effects over time
interaction_coeffs=[
    coefficients['esg_Year_2020_interaction'],
    coefficients['esg_Year_2021_interaction'],
    coefficients['esg_Year_2022_interaction'],
    coefficients['esg_Year_2023_interaction']
]

years=[2020, 2021, 2022, 2023]

plt.figure(figsize=(10, 6))
plt.plot(years,interaction_coeffs,marker='o',color='b',label='ESG=')
plt.title('Impact of ESG on TQ Over Time')
plt.xlabel('Year')
plt.ylabel('Coefficient')
plt.axhline(0,color='red',linestyle='--',linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Hausman Test for selecting fe or re
from scipy.stats import chi2
from linearmodels.panel import PanelOLS
yh=panel_data['TQ']
xh=panel_data[['esg','age','lev','beta']]
xh=sm.add_constant(xh)

fe_model=PanelOLS(yh,xh,time_effects=True).fit()

re_model=RandomEffects(yh,xh).fit()

b_fe=fe_model.params
b_re=re_model.params

diff=b_fe-b_re

cov_fe=fe_model.cov
cov_re=re_model.cov

cov_diff=cov_fe-cov_re

hausman_stat=np.dot(np.dot(diff.T,np.linalg.inv(cov_diff)),diff)

df=len(diff)
p_value=1-chi2.cdf(hausman_stat,df)

print(f'Hausman statistic:{hausman_stat}')
print(f'P value:{p_value}')

if p_value<0.05:
    print("Choose fe")
else:
    print("Choose re")
```
